```python
import pandas as pd
import re
import matplotlib.pyplot as plt
from scipy import stats
```

```python
dog_df = pd.read_csv('dog_rates_tweets.csv', parse_dates=[1])
```

```python
ratings = dog_df['text'].str.extract(r'(\d+(\.\d+)?)/10', expand=False)
ratings = pd.to_numeric(ratings[0])
dog_df['rating'] = ratings
dog_df = dog_df.dropna()
dog_df = dog_df[dog_df['rating'] <= 25.0]
dog_df
```

```python
plt.figure(figsize=(20,10))
plt.scatter(dog_df['created_at'].values, dog_df['rating'])
plt.xlabel("Date")
plt.ylabel("Rating")
plt.show
```

```python
def to_timestamp(ca):
    return ca.timestamp()   
```

```python
dog_df['timestamp'] = dog_df['created_at'].apply(to_timestamp)
```

```python
fit = stats.linregress(dog_df['timestamp'], dog_df['rating'] )
```

```python
dog_df['prediction'] = (fit.slope)*dog_df['timestamp'] + fit.intercept
```

```python
dog_df
```

```python
fit.slope,fit.intercept
```

```python
plt.xticks(rotation=25)
plt.plot(dog_df['created_at'].values, dog_df['rating'], 'b.', alpha=0.5)
plt.plot(dog_df['created_at'], dog_df['timestamp']*fit.slope + fit.intercept, 'r-', linewidth=3)
plt.show()
```
